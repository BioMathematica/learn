# Exploratory, potentially-outdated notebook exploring feature attributes for blacklisting

```python
import pandas
from statsmodels.sandbox.stats.multicomp import multipletests
```

```python
auroc_df = pandas.read_table('./data/feature-performance/auroc.tsv')
reject, pvals_corrected, alphacSidak, alphacBonf = multipletests(auroc_df.pval_delta_auroc, method='fdr_bh')
auroc_df['fdr_delta_auroc'] = pvals_corrected
auroc_df.head()
```

```python
deadlist = [
    'CtDaGiGaD', 'CtDaGaDrD', 'CtDpSpDrD', 'CtDlAlDrD',
    'CtDlAdGaD', 'CtDrDpSpD', 'CtDrDaGaD', 'CtDaGbCtD', 'CbGr>GdCtD',
]
auroc_df.query("metapath in @deadlist")
```

```python
auroc_df.query("rdwpc_auroc > 0.525")
```

```python
auroc_df.query("rdwpc_auroc < 0.4 and pdwpc_auroc > 0.65")
```

```python
auroc_df.sort_values('rdwpc_auroc')[:100]
```

```python
blacklist_df = auroc_df.query("rdwpc_auroc < 0.5")
blacklist_df = blacklist_df.loc[blacklist_df.metapath.str.contains(r'[CD][t][CD]'), :]
blacklist_df['feature'] = 'dwpc_' + blacklist_df['metapath']
blacklist_df.head()
```

```python
blacklist_df[['metapath', 'feature']].to_csv('./selection/blacklist.tsv', index=False, sep='\t')
```

```python
whitelist_df = auroc_df.query(
    "rdwpc_auroc > 0.55 and delta_auroc > 0 and fdr_delta_auroc < 0.05 and pdwpc_primary_auroc > 0.5").copy()
whitelist_df['feature'] = 'dwpc_' + whitelist_df['metapath']
whitelist_df.head()
whitelist_df.metapath.map(lambda x: x[:3]).value_counts()
```

```python
whitelist_df[['metapath', 'feature']].to_csv('./selection/whitelist.tsv', index=False, sep='\t')
```
